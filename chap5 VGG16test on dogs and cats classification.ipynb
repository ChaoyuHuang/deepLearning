### BACKGROUND 用vgg16测试一下猫狗识别

```python
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
import numpy as np
import pandas as pd
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
model = VGG16()
model.summary()
```

```python
import os
FILE_PATH = './dog-cat-full-dataset-master/data/train/dogs/'
```

```python
for file in os.listdir(FILE_PATH):
    full_path = FILE_PATH + file
    image = load_img(full_path, target_size=(224, 224))
    image = img_to_array(image)
    image = image.reshape(1, image.shape[0], image.shape[1], image.shape[2])
    image = preprocess_input(image)
    y_pred = model.predict(image)
    label = decode_predictions(y_pred, top=1)
    print(label)
    print()
```

## 测试用Imagegenerator feed数据

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(rescale=1/225.)
```

```python
train_genertator = datagen.flow_from_directory(directory='./dog-cat-full-dataset-master/data/train/',
                                              target_size=(224, 224),
                                              classes=['cats', 'dogs'],
                                              class_mode='binary',
                                              batch_size=20)
```

```python
pred = model.predict_generator(generator=train_genertator, verbose=1)
```

```python
predicted_class_indices = np.argmax(pred, axis=1)
```
